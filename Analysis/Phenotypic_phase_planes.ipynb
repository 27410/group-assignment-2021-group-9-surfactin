### Phenotypic Phase Plane Analysis

```python
from cobra.flux_analysis.phenotype_phase_plane import production_envelope
```

#### PPP for biomass on glucose 

```python
ppp_biomass = production_envelope(model,
                    reactions=[model.reactions.EX_glc__D_e],
                    objective=model.reactions.BIOMASS_BS_10)
ppp_biomass
```

```python
ppp_biomass.plot(x='EX_glc__D_e', y='flux_maximum',title="Biomass flux on glucose",ylabel="Biomass") #Plotting the ppp
```

#### PPP for surfactin on glucose

```python
ppp_surfactin = production_envelope(model,
                    reactions=[model.reactions.EX_glc__D_e],
                    objective=model.reactions.DM_surfactin_c)
ppp_surfactin
```

```python
ppp_surfactin.plot(x='EX_glc__D_e',y='flux_maximum',title="Surfactin flux on glucose",ylabel="Surfactin") #plotting the ppp
```

#### Testing surfactin production for aerobic vs anaerobic conditions

```python
medium = model.medium
with model:
    medium['EX_o2_e'] = 0 #removing oxygen from medium
    model.medium = medium #defining medium without O
    solution = model.optimize() #optimizing this solution
    print(solution.fluxes['BIOMASS_BS_10']) 
```

Not viable, our organism cannot grow without oxygen.
